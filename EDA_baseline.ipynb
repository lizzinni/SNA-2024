### поиск похожих мест по меткам

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
```

```python
X = items[all_feature_columns]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

cosine_sim = cosine_similarity(X_scaled)
```

```python
def get_similar_places(place_name, n=5):

    idx = items[items['name'] == place_name].index[0]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_places = sim_scores[1:n+1]
    similar_places = [(items['name'].iloc[i[0]], i[1]) for i in sim_places]

    return similar_places
```

```python
import random

random_places = random.sample(list(items['name']), 5)

for place in random_places:
    similar_places = get_similar_places(place, n=5)
    print(f"Похожие места для '{place}':")
    for similar in similar_places:
        print(f"- {similar[0]}")
    print()
```

### ищем похожих юзеров

```python
merged_reviews = reviews.merge(items[['detail_id', 'name']], on='detail_id', how='left')

user_places_with_ratings = merged_reviews.groupby('profile_url').agg({'name': list, 'mark': list}).reset_index()

user_places_with_ratings.columns = ['profile_url', 'places_rated', 'ratings']

user_places_with_ratings.head()
```

```python
users.columns
```

```python
ratings_df = pd.DataFrame({
    'profile_url': user_places_with_ratings['profile_url'].repeat(user_places_with_ratings['places_rated'].str.len()),
    'place': [place for places in user_places_with_ratings['places_rated'] for place in places],
    'rating': [rating for ratings in user_places_with_ratings['ratings'] for rating in ratings]
})

user_ratings = ratings_df.pivot_table(index='profile_url', columns='place', values='rating', fill_value=0)
user_ratings
```

```python
def find_similar_users(user_ratings, target_user, n=5):
    target_user_ratings = user_ratings.loc[target_user].values.reshape(1, -1)
    similarity_scores = cosine_similarity(user_ratings.drop(target_user).values, target_user_ratings)
    similar_users_idx = similarity_scores.flatten().argsort()[-n:][::-1]
    similar_users = user_ratings.drop(target_user).index[similar_users_idx]
    return similar_users
```

```python
def recommend_places(target_user, similar_users, user_ratings, n=5):
    similar_users_ratings = user_ratings.loc[similar_users]

    target_user_rated_places = user_ratings.loc[target_user].to_numpy() != 0
    all_places = user_ratings.columns
    places_to_recommend = [place for place in all_places if not target_user_rated_places[all_places.get_loc(place)]]

    recommended_places = {}
    for place in places_to_recommend:
        avg_rating = similar_users_ratings[place].mean()
        recommended_places[place] = avg_rating

    sorted_recommended_places = sorted(recommended_places.items(), key=lambda x: x[1], reverse=True)
    return [place for place, _ in sorted_recommended_places[:n]]
```

```python
# Находим похожих пользователей для '/Profile/Young_traveller_102'
similar_users = find_similar_users(user_ratings, '/Profile/Young_traveller_102', n=5)

# Рекомендуем места для '/Profile/Young_traveller_102' на основе оценок похожих пользователей
recommended_places = recommend_places('/Profile/Young_traveller_102', similar_users, user_ratings)

print(f"Рекомендованные места для '/Profile/Young_traveller_102': {recommended_places}")
```

### граф похожих пользователей

```python
import networkx as nx
```

```python
user_ratings
```

```python
user_ratings.set_index('profile_url', inplace=True)

similarity_matrix = cosine_similarity(user_ratings.fillna(0))

similarity_df = pd.DataFrame(similarity_matrix, index=user_ratings.index, columns=user_ratings.index)

G = nx.Graph()
```

```python
threshold = 0.5

for user1 in similarity_df.index:
    for user2 in similarity_df.columns:
        if user1 != user2 and similarity_df.loc[user1, user2] > threshold:
            G.add_edge(user1, user2, weight=similarity_df.loc[user1, user2])

edge_weights = nx.get_edge_attributes(G, 'weight')
max_weight = max(edge_weights.values()) if edge_weights else 0

plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels=False, node_color='pink', node_size=20)

nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')

highlight_edges = [(u, v) for u, v, w in G.edges(data=True) if w['weight'] >= max_weight * 0.5]
nx.draw_networkx_edges(G, pos, edgelist=highlight_edges, edge_color='red', width=1)

plt.title("Граф сходства пользователей (подсвечены тяжелые ребра)")
plt.show()
```

```python
def get_top_similar_users(user, G, top_n=5):
    neighbors = [(neighbor, G[user][neighbor]['weight']) for neighbor in G.neighbors(user)]
    top_similar_users = sorted(neighbors, key=lambda x: x[1], reverse=True)[:top_n]
    return [user for user, _ in top_similar_users]

user = '/Profile/070165'
similar_users = get_top_similar_users(user, G)
print(f"Похожие пользователи для {user}: {similar_users}")
```

```python
def recommend_places(user, similar_users, user_ratings):
    user_places = set(user_ratings.loc[user][user_ratings.loc[user] > 0].index)

    recommended_places = set()
    for similar_user in similar_users:
        similar_user_places = set(user_ratings.loc[similar_user][user_ratings.loc[similar_user] > 0].index)
        recommended_places.update(similar_user_places - user_places)

    return list(recommended_places)

recommended_places = recommend_places(user, similar_users, user_ratings)
print(f"Рекомендованные места для {user}: {recommended_places}")
```

### граф рекомендаций

```python
B = nx.Graph()

user_nodes = user_places_with_ratings['profile_url']
place_nodes = set([place for places in user_places_with_ratings['places_rated'] for place in places])

B.add_nodes_from(user_nodes, bipartite=0)
B.add_nodes_from(place_nodes, bipartite=1)

for _, row in user_places_with_ratings.iterrows():
    user = row['profile_url']
    for place, rating in zip(row['places_rated'], row['ratings']):
        B.add_edge(user, place, weight=rating)
```

```python
user_projection = nx.bipartite.weighted_projected_graph(B, user_nodes)

plt.figure(figsize=(12, 9))
pos = nx.spring_layout(B, seed=42)
nx.draw(B, pos, with_labels=False, node_size=100, node_color='lightgreen', edge_color='gray')
plt.title("Граф рекомендаций (двудольный)")
plt.show()
```

```python
def recommend_places_for_user(user, B, top_n=5):
    user_neighbors = set(B.neighbors(user))
    all_places = {node for node, data in B.nodes(data=True) if data.get('bipartite') == 1}
    unseen_places = all_places - user_neighbors

    recommendations = []
    for place in unseen_places:
        scores = [B[user, place]['weight'] for user in user_projection.neighbors(user) if B.has_edge(user, place)]
        if scores:
            avg_score = sum(scores) / len(scores)
            recommendations.append((place, avg_score))

    recommendations.sort(key=lambda x: x[1], reverse=True)

    return [place for place, _ in recommendations[:top_n]]
```

```python
recommended_places = recommend_places_for_user('/Profile/070165', B)
print(f"Рекомендованные места для '/Profile/070165': {recommended_places}")
```

### random walk

```python
user_ratings
```

```python
G = nx.Graph()

users = user_ratings.index
places = user_ratings.columns

G.add_nodes_from(users, bipartite=0)
G.add_nodes_from(places, bipartite=1)

for user in users:
    for place in places:
        if user_ratings.loc[user, place] > 0:
            G.add_edge(user, place, weight=user_ratings.loc[user, place])
```

```python
def random_walk(graph, start_node, walk_length=100):
    current_node = start_node
    visited = {current_node: 1}

    for _ in range(walk_length):
        neighbors = list(graph.neighbors(current_node))
        if not neighbors:
            break
        next_node = random.choice(neighbors)
        visited[next_node] = visited.get(next_node, 0) + 1
        current_node = next_node

    return visited
```

```python
walk_results = {}
for user in users:
    walk_results[user] = random_walk(G, user, walk_length=50)

for user, result in walk_results.items():
    print(f"Результаты случайного блуждания для {user}:")
    for node, count in result.items():
        print(f"  {node}: {count}")
    break
```

```python
for user, result in walk_results.items():
    recommended_places = sorted(result.items(), key=lambda x: x[1], reverse=True)
    print(f"Рекомендации для {user}:")
    i = 0
    for place, count in recommended_places:
      if i == 5:
        break
      print(f"  {place}: {count}")
      i += 1
    break
```
