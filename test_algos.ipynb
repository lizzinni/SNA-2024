```python
def find_similar_users(user_ratings, target_user, n=5):
    target_user_ratings = user_ratings.loc[target_user].values.reshape(1, -1)
    similarity_scores = cosine_similarity(user_ratings.drop(target_user).values, target_user_ratings)
    similar_users_idx = similarity_scores.flatten().argsort()[-n:][::-1]
    similar_users = user_ratings.drop(target_user).index[similar_users_idx]
    return similar_users
```

```python
def recommend_places(target_user, similar_users, user_ratings, n=5):
    similar_users_ratings = user_ratings.loc[similar_users]

    target_user_rated_places = user_ratings.loc[target_user].to_numpy() != 0
    all_places = user_ratings.columns
    places_to_recommend = [place for place in all_places if not target_user_rated_places[all_places.get_loc(place)]]

    recommended_places = {}
    for place in places_to_recommend:
        avg_rating = similar_users_ratings[place].mean()
        recommended_places[place] = avg_rating

    sorted_recommended_places = sorted(recommended_places.items(), key=lambda x: x[1], reverse=True)
    return [place for place, _ in sorted_recommended_places[:n]]
```

### Граф рекомендаций

```python
B = nx.Graph()

user_nodes = user_places_with_ratings['profile_url']
place_nodes = set([place for places in user_places_with_ratings['places_rated'] for place in places])

B.add_nodes_from(user_nodes, bipartite=0)
B.add_nodes_from(place_nodes, bipartite=1)

for _, row in user_places_with_ratings.iterrows():
    user = row['profile_url']
    for place, rating in zip(row['places_rated'], row['ratings']):
        B.add_edge(user, place, weight=rating)

user_projection = nx.bipartite.weighted_projected_graph(B, user_nodes)

plt.figure(figsize=(18, 12)) 
pos = nx.spring_layout(B, seed=42, k=0.6) 

user_pos = {node: pos[node] for node in user_nodes}
place_pos = {node: pos[node] for node in place_nodes}

nx.draw_networkx_nodes(B, pos, nodelist=user_nodes, node_color='purple', node_size=50) 

nx.draw_networkx_nodes(B, pos, nodelist=place_nodes, node_color='pink', node_size=30) 

weights = [B[u][v]['weight'] for u, v in B.edges()]
nx.draw_networkx_edges(B, pos, edge_color='black', width=[w * 0.2 for w in weights])  

plt.title("Граф рекомендаций (двудольный)", fontsize=14)
plt.axis('off')
plt.show()
```

```python
def recommend_places_for_user(user, B, user_projection, items_df, top_n=5):
    if user not in B:
        print(f"Пользователь '{user}' отсутствует в графе.")
        return []

    user_neighbors = set(B.neighbors(user))

    all_places = {node for node, data in B.nodes(data=True) if data.get('bipartite') == 1}

    unseen_places = all_places - user_neighbors

    recommendations = []
    for place in unseen_places:
        scores = []
        for neighbor in user_projection.neighbors(user):
            if B.has_edge(neighbor, place):
                scores.append(B[neighbor][place]['weight'])
        if scores:
            avg_score = sum(scores) / len(scores)
            recommendations.append((place, avg_score))

    recommendations.sort(key=lambda x: x[1], reverse=True)

    recommended_places = [place for place, _ in recommendations[:top_n]]
    detail_ids = items.loc[items['name'].isin(recommended_places), 'detail_id'].tolist()

    return detail_ids


user_recommendations = {}

for user in tqdm(test_users):
    recommended_detail_ids = recommend_places_for_user(user, B, user_projection, items)
    user_recommendations[user] = recommended_detail_ids


with open('graph_recommendations_.json', 'w', encoding='utf-8') as f:
    json.dump(user_recommendations, f, ensure_ascii=False, indent=4)

print("Рекомендации сохранены в файл graph_recommendations_.json")
```

```python
G = nx.Graph()

users = user_ratings.index
places = user_ratings.columns

G.add_nodes_from(users, bipartite=0)
G.add_nodes_from(places, bipartite=1)

for user in users:
    for place in places:
        if user_ratings.loc[user, place] > 0:
            G.add_edge(user, place, weight=user_ratings.loc[user, place])
```

```python
def random_walk(graph, start_node, walk_length=1000):
    current_node = start_node
    visited = {current_node: 1}

    for _ in range(walk_length):
        neighbors = list(graph.neighbors(current_node))
        if not neighbors:
            break
        next_node = random.choice(neighbors)
        visited[next_node] = visited.get(next_node, 0) + 1
        current_node = next_node

    return visited
```

```python
def get_top_places_with_detail_ids(walk_results, place_nodes, place_to_detail_id, top_n=5):
    place_counts = {node: count for node, count in walk_results.items() if node in place_nodes}

    sorted_places = sorted(place_counts.items(), key=lambda x: x[1], reverse=True)

    return [place_to_detail_id[place] for place, _ in sorted_places[:top_n] if place in place_to_detail_id]

place_to_detail_id = dict(zip(items['name'], items['detail_id']))

walk_results = {}
for user in test_users:
    walk_results[user] = random_walk(G, user, walk_length=50)

user_recommendations = {}

for user, result in walk_results.items():
    top_detail_ids = get_top_places_with_detail_ids(result, place_nodes, place_to_detail_id)
    user_recommendations[user] = top_detail_ids

with open('random_walk_recommendations.json', 'w', encoding='utf-8') as f:
    json.dump(user_recommendations, f, ensure_ascii=False, indent=4)

print("Рекомендации сохранены в файл random_walk_recommendations.json")
```
