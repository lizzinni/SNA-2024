```python
# всего тегов
len(all_feature_columns)
```

```python
# среднее количество меток для каждой достопримечательности

items['num_tags'] = items[all_feature_columns].sum(axis=1)

average_tags_per_place = items['num_tags'].mean()
min_tags_per_place = items['num_tags'].min()
max_tags_per_place = items['num_tags'].max()

print(f"Среднее количество меток для каждой достопримечательности: {average_tags_per_place:.2f}")
print(f"Минимальное количество меток для достопримечательности: {min_tags_per_place}")
print(f"Максимальное количество меток для достопримечательности: {max_tags_per_place}")
```

```python
place_with_max_tags = items.loc[items['num_tags'].idxmax()]

place_with_min_tags = items.loc[items['num_tags'].idxmin()]

print("Место с максимальным количеством меток:")
print(place_with_max_tags[['name', 'num_tags']])

print("\nМесто с минимальным количеством меток:")
print(place_with_min_tags[['name', 'num_tags']])
```
