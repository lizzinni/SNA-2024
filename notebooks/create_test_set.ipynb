```python
reviews = pd.read_csv('reviews.csv')
reviews.head()
```

```python
users = pd.read_csv('users.csv')
users.shape
```

```python
reviews_sorted = reviews.sort_values(by=['profile_url', 'date'], ascending=False)
reviews_sorted
```

```python
top_profiles = reviews_sorted['profile_url'].value_counts().index[:200]

selected_reviews = pd.DataFrame()

for profile in top_profiles:
    selected_reviews = pd.concat([selected_reviews, reviews_sorted[reviews_sorted['profile_url'] == profile].head(5)])

reviews_sorted = reviews_sorted[~reviews_sorted.index.isin(selected_reviews.index)]
```

```python
reviews_sorted.to_csv('reviews_test.csv', index=False)
```

```python
selected_reviews.to_csv('reviews_test_true.csv', index=False)
```

```python
unique_profile_urls = selected_reviews['profile_url'].unique()

with open('test_profile_urls.txt', 'w') as f:
    for url in unique_profile_urls:
        f.write(f"{url}\n")
```
